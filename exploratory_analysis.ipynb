```python
import pandas as pd
import numpy as np 
from sqlalchemy import create_engine
from scipy.stats import pearsonr
```

```python
# creating fire data
engine = create_engine("sqlite:///data.sqlite")
connection = engine.connect()
```

```python
average_fire_df = pd.read_sql("SELECT FIRE_YEAR, AVG(FIRE_SIZE) AS average_fire_size FROM FIRES GROUP BY FIRE_YEAR", connection)
```

```python
average_fire_df.head(30)
```

```python
x = average_fire_df['FIRE_YEAR']
y = average_fire_df['average_fire_size']
```

```python
corr, p = pearsonr(x,y)
print(f"the correaltion: {corr}")
print(f"p-value: {p}")
```

```python
count_df = pd.read_sql("SELECT FIRE_YEAR, COUNT(FOD_ID) AS COUNT_FIRES FROM FIRES GROUP BY FIRE_YEAR", connection)
count_df.head(30)
```

```python
count_x = count_df["FIRE_YEAR"]
count_y = count_df['COUNT_FIRES']
```

```python
corr, p = pearsonr(count_x, count_y)
print(corr)
print(p)
```

```python
cali_df = pd.read_sql("SELECT FIRE_YEAR, COUNT(FOD_ID) AS COUNT_FIRES, AVG(FIRE_SIZE) AS AVG_FIRE_SIZE FROM FIRES WHERE STATE='CA' GROUP BY FIRE_YEAR", connection)
cali_df.head(30)
```

```python
year = cali_df["FIRE_YEAR"]
count = cali_df["COUNT_FIRES"]
size = cali_df["AVG_FIRE_SIZE"]
```

```python
corr, p = pearsonr(year, count)
print(corr)
print(p)
```

```python
corr, p = pearsonr(year, size)
print(corr)
print(p)
```

```python
pop2020 = population_df.loc[(population_df["Year"] == 2020), :]
pop2020 = pop2020.drop(679, axis="index")
pop2020 = pop2020.drop(680, axis="index")
pop2020 = pop2020.drop(681, axis="index")
pop2020 = pop2020.drop(682, axis="index")
pop2020 = pop2020.drop(683, axis="index")
pop2020 = pop2020.drop(635, axis="index")
pop2020 = pop2020.drop(666, axis = "index")
pop2020["Resident Population Density"] = pop2020["Resident Population Density"].str.replace(',', '')
pop2020["Resident Population Density"] = pop2020["Resident Population Density"].astype(float)
pop2020["STATE"] = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "FL", "GA", "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
pop2020
```

```python
fire2020 = pd.read_sql("SELECT STATE, COUNT(FIRE_SIZE) AS TOTAL_ACRES FROM FIRES WHERE FIRE_YEAR = 2020 GROUP BY STATE", connection)
fire2020 = fire2020.drop(7, axis="index")
```

```python
combined_df = pop2020.merge(fire2020, on = "STATE")
combined_df
```
